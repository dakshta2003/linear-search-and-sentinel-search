```python
def search(list,key):
  for i in range(len(list)):
    if key==list[i]:
       print("roll number is found")
       break
  else:
    print("roll number is not found")
list=[22,3,55,23,11,33,44,88,42,82,8]
print(list)
key=int(input("enter the roll number"))
search(list,key)
```
